<p style="font-size:18px;"> Problem Solving</p>
<p style="font-size:16px;">Assume you open your ice cream shop, there are only two types of ice cream, vanilla and strawberry.</br>
When a box of ice cream is sold, you will get the benefit for $2 for vanilla ice cream and $3 for strawberry ice cream.</br>
To make the ice cream, the fresh milk is required. To make a box of vanilla ice cream requires 0.5 liter and strawberry</br> ice cream requires 0.2 liter. 
You daily order 10 liters of fresh milk.To promote your ice cream shop,</br> 
you give a doll for each ice cream box. The number of dolls, that you can give to customers, is 30 dolls per day.</br> 
So, how many boxes of vanilla ice cream and strawberry ice cream that you would like to produce to get maximum profit ? </p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
import math
```

<h2 style="font-weight:bold; text-decoration:underline;">  Decision variable </h2>
<p style="font-size:18px;">x <sub>i</sub> : the number of ice cream type i</p>
<p style="margin-left:2.5%; margin-right:0%;font-size:16px;"> 
x<sub>1</sub> : Strawberry ice cream box</br>
x<sub>2</sub> : Vanilla ice cream box

</p>
<hr>
</br>

<h2 style="font-weight:bold; text-decoration:underline;"> Objective </h2>
<p style="font-size:18px;">MAX PROFIT: 3x<sub>1</sub> + 2x<sub>2</sub></p>
<hr>
</br>

<h2 style="font-weight:bold; text-decoration:underline;"> Constraint </h2>
<p style="font-size:18px;">The quantity of milk: 0.5x<sub>1</sub> + 0.2x<sub>2</sub> &#8804 10</p>
<p style="font-size:18px;">The numberof ice cream box: x<sub>1</sub> + x<sub>2</sub> &#8804 30</p>
<p style="font-size:18px;">x<sub>1</sub>, x<sub>2</sub> &#8805 0</p>

<hr>
</br>




```python
# 2 lines graph cut point
co_eff = np.array([[0.5, 0.2], [1., 1.]])
b= np.array([10, 30])
x = solve(co_eff, b)
print(x) # x should be integer because number of box can't be float
x[0] = math.floor(x[0])
x[1] = math.floor(x[1])
print(x)
```

```python
# Create Graph to Visualize overall
x1 = np.array(range(31))

y1 = 30 - x1
y2 = (10 - 0.5*x1)/0.2

plt.plot(x1, y1, linestyle="none")
plt.plot(x1, y2, linestyle="none")

plt.xlim(0, 40)
plt.ylim(0, 60)

plt.fill_between(x1, y1, hatch='/////')
plt.fill_between(x1, y1, y2, color='#e8b787')
plt.text(0, 30, "A") # y-intercept
plt.text(13.33333333, 16.66666667, "X") # graph cut point
plt.text(20, 0, "B") # x-intercept

# set x/y labels
plt.xlabel("# Strawberry Ice-cream Box")
plt.ylabel("# Vanilla ice-cream box")
plt.title("Constraint")

plt.show()
```

```python
A = [0., 30.]
B = [20., 0.]
def price_profit(matrix):
    profit = 3*matrix[0] + 2*matrix[1]
    return profit
profit1 = price_profit(A)
profit2 = price_profit(x)
profit3 = price_profit(B)

list_profit = [profit1, profit2, profit3]
print(f"the hightest profit = {max(list_profit)} USD.")
    
```
