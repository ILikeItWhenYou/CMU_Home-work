### 1.) Which region sold <i style = "color:#a83232;">the largest</i> amount of avocado ?
### In this region, where the biggest lot of sold avocado came from (4046, 4225, 4770) ?
<p style="color:MediumSeaGreen;"><b>Answer: From the result as below, "TotalUS region" sold the largest amount of avocado
and the biggest lot is "4046"</b></p>


```python
df_region= df.groupby('region', sort = False)[['Total Volume', '4046', '4225', '4770']].sum()
df_region_asc = df_region.sort_values(by='Total Volume', ascending=False)
df_region_asc.head()
```

### 2.) Which region sold <i style="color:#a83232;">the smallest</i> amount of avocado ?
### In this region, where the biggest lot of sold avocado came from (4046, 4225, 4770) ?
<p style="color:MediumSeaGreen;"><b>Answer: From the result as below, "Syracuse region" sold the smallest amount of avocado
and the biggest lot is "4225"</b></p>

```python
df_region= df.groupby('region', sort = False)[['Total Volume', '4046', '4225', '4770']].sum()
df_region_dsc = df_region.sort_values(by='Total Volume')
df_region_dsc.head()
```

### 3.) Which region sold <i style="color:#a83232;">the highest</i> amount of avocado in average?
<p style="color:MediumSeaGreen;"><b>Answer: "TotalUS region" has the highest average amount of selling volume of avocado</b></p>

```python
df_region_avg = df.groupby('region', sort = False)[['Total Volume']].mean()
df_region_avg = df_region_avg.sort_values('Total Volume', ascending=False)
df_region_avg.head()
```

### 4.) Find the total amount of income (Avg_Price*Total_Volume) of each region.?

```python
df['total_income'] = df['AveragePrice'] * df['Total Volume']
df_income = df[['region', 'total_income']].groupby('region', sort=False).sum('total_income')
df_income
```

### 5) Let AVOCADO  Average Weight : 4046 => 4 ounces, 4225 => 9 ounces, 4770 => 12 ounces
###    Find the number of sold avocadoes by region ?
###    Which region sold the largest number of avocadoes ?
<p style="color:MediumSeaGreen;"><b>Answer: The number of sold avocadoes are shown as the table below. And the region where
sold the largest number of avocadoes is "TotalUS"</b></p>

```python
# total number of Avocado in 4046 
df['num_4046'] = df['4046']/4
# total number of Avocado in 4225 
df['num_4225'] = df['4225']/9
# total number of Avocado in 4770
df['num_4770'] = df['4770']/12
df['number_sold'] = df['num_4046'] + df['num_4225'] + df['num_4770']

df_num_region = df.groupby('region', sort=False)[['region', 'number_sold']].sum()
df_num_region.sort_values(by='number_sold', ascending = False)
```

### 6) Normally, the customers buy the avocados by unit or in a bags ?
<p style="color:MediumSeaGreen;"><b>Answer: From the result shows that, the customers usually buy the avocados by unit</b></p>

```python
# Total volume = 4046 + 4225 + 4470 + Total bags
df_all_unit= df['4046'] + df['4225'] + df['4770']
df_all_unit.sum() > df['Total Bags'].sum()
```

## Optional

### 7) Which month is the best time to sell avocado?
<p style="color:MediumSeaGreen;"><b>Answer: In Febuary is the best time to sell avocado</b></p>

```python
# Change Date time format to Y-M, then group them tgt.
df['Date_new'] = pd.to_datetime(df['Date']).dt.strftime('%m')

# Compare number of sold avocado for each month(for all years)
df_month = df.groupby('Date_new')[['number_sold']].sum()

# Sort number of sold to see the highest value
df_month.sort_values('number_sold', ascending=False)
```
