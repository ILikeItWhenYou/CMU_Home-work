*Exercise: Group by method*

```python
# Group by data using rank
df_rank = df.groupby(['rank'])

# Calculate mean for each numeric column per each group
df_rank.mean()
```

```python
# Once groupby object
df.groupby('rank')[['salary']].mean()
```

```python
# group keys are soerted by default. try sort = False
df.groupby(['rank'], sort = False)[['salary']].mean()
```

*Exercise: filtering*

```python
# salary > 120k usd
df_sub = df[df['salary'] > 120000]
df_sub.head()
```

```python
# select only female professors
df_f = df[df['sex'] == 'Female']
df_f.head()
```
